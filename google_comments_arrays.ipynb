```python
import pandas
import json
from itertools import chain
```

```python
f = open('Z:/- RESPONSABLE DIGITAL/SCRIPTS/google_comments/google_comments/media/maison_chapon.json', mode='r', encoding='utf-8')
data = json.load(f)
f.close()
```

```python
comments = list(chain(*data))
```

```python
comments
```

```python
df = pandas.DataFrame(comments)
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=['review_id'])
```

```python
df = df[['text', 'rating', 'period']]
```

```python
df = df.sort_values('text')
```

```python
df.to_csv('maison_chapon.csv', encoding='utf-8', index=False, sep=';')
```
