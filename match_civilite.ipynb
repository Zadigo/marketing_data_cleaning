```python
import pandas
from marketing_data_cleaning import DATA_FOLDER_PATH
```

```python
df = pandas.read_csv(DATA_FOLDER_PATH / 'prenoms.csv', encoding='utf-8', sep=';')
```

```python
df = df.rename(columns={
    '01_prenom': 'prenom',
    '02_genre': 'genre',
    '03_langage': 'langage',
    '04_fréquence': 'fréquence'
})
```

```python
df = df[['prenom', 'genre']]
```

```python
df.head()
```

```python
girl_names = df[df['genre'] == 'f']
boy_names = df[df['genre'] == 'm']
mixed_naems = df[df['genre'] == 'm,f']
```

```python
girl_names.describe()
```

```python
initial_db = pandas.read_csv(DATA_FOLDER_PATH / 'db/v1.csv', encoding='utf-8')
```

```python
def name_checker(value):
    value = str(value).lower()

    result = girl_names[girl_names['prenom'] == value].empty
    if not result:
        return 'G'
    
    result = boy_names[boy_names['prenom'] == value].empty
    if not result:
        return 'B'
    return 'A'


initial_db['gender'] = initial_db['Prénom'].apply(name_checker)
```
