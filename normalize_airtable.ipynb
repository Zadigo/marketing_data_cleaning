```python
import airtable
import os
from marketing_data_cleaning import DATA_FOLDER_PATH
```

```python
os.getenv('LAUREN_BASE_ID')
```

```python
instance = airtable.Airtable(
    os.getenv('LAUREN_BASE_ID'),
    os.getenv('LAUREN_TABLE_ID'),
    os.getenv('AIRTABLE_API_KEY')
)
```

```python
data = instance.get_all()
```

```python
data
```

```python
records_to_update = []
for item in data:
    fields = item.get('fields', {})
    updated_record = {
        'id': item['id'],
        'fields': {
            'Nom': fields.get('Nom', '').lower().title(),
            'Prénom': fields.get('Prénom', '').lower().title(),
            'Entreprise': fields.get('Entreprise', '').upper()
        }
    }
    records_to_update.append(updated_record)
```

```python
instance.batch_update(records_to_update)
```
