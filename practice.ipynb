```python
import pandas
import requests
from matplotlib import pyplot
```

```python
url = 'https://data.opendatasoft.com/api/explore/v2.1/catalog/datasets/prix-carburants-fichier-instantane-test-ods-copie@opendatamef/records?limit=20'
response = requests.get(url, headers={'content-type': 'application/json'})
```

```python
data = response.json()['results']
```

```python
df = pandas.DataFrame(data)
```

```python
df = df[['cp', 'pop', 'ville', 'prix_id', 'prix_valeur', 'dep_name']]
```

```python
df = df.sort_values('prix_valeur')
```

```python
fig = pyplot.figure(figsize=(200, 200))
fix, axes = pyplot.subplots(figsize=(15, 6))
axes.set_xlabel('Region')
axes.set_ylabel('Count')
axes.set_title("Prices")
axes.tick_params(which='major', width=1.00, length=5)
# axes.text(20, 35, 'Some text')
# axes.annotate('Something', xy=[30, 40], xytext=[14, 31], arrowprops={'facecolor': 'black', 'shrink': 0.05})
# axes.set_xticks([0, 30, 70, 100])

# for i in range(6):
#     axes.text(i, df['prix_valeur'][i], df['prix_valeur'][i])

axes.legend('Prix carburant')
# axes.plot(x, y, 'o', label='words')
axes.bar(df['prix_id'], df['prix_valeur'], color='b', width=0.4)
fig.show()
```
